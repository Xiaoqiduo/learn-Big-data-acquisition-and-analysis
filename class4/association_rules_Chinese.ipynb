```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

selected_courses = [['C','数据结构','算法'],
          ['Java','数据结构','数据库'],
          ['C','Java','数据结构','数据库'],
          ['Java','数据库']]
```

数据转换为OneHot

```python
from mlxtend.preprocessing import TransactionEncoder
oht = TransactionEncoder()
oht_ary = oht.fit(selected_courses).transform(selected_courses)
oht_ary = oht_ary.astype("int")
df = pd.DataFrame(oht_ary, columns=oht.columns_)
df
```

```python
apriori(df, min_support=0.6)
```

```python
frequent_itemsets=apriori(df, min_support=0.5, use_colnames=True)
frequent_itemsets
```

```python
rules_lift = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules_lift
```

```python
rules_confidence = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.6)
rules_confidence
```

```python
rules_confidence[ (rules_confidence['lift'] >= 1) &
       (rules_confidence['confidence'] >= 0.8) ]
```

```python
rules_confidence["antecedent_len"] = rules_confidence["antecedents"].apply(lambda x: len(x))
rules_confidence
```

```python
rules_confidence[ (rules_confidence['antecedent_len'] >= 2) &
       (rules_confidence['confidence'] > 0.8) &
       (rules_confidence['lift'] > 1) ]
```
